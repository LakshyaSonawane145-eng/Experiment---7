```python
# Print i as long as i is less than 6:
i = 1
while i < 6:
  print(i)
  i +=1
```

```python
# Print numbers from 1 to N
n = int(input("Enter N: "))
i = 1
while i <= n:
  print(i)
  i += 1
```

```python
#Factorial of a number
n = int(input("Enter N: "))
fact = 1
while n > 0:
  fact *= n
  n -= 1
print("Factorial =" ,fact)
```

```python
#Fibonacci Series using while loop
n = int(input("Enter number of terms:"))
a = 0
b = 1
i = 1
while i <= n:
  print(a,end = " ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
limit = int(input("Enter limit: "))
a = 0
b = 1
while a <= limit:
    print(a, end=" ")
    a, b = b, a + b
```
