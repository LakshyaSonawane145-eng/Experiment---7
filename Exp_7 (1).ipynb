```python
# Print numbers from 1 to N
n = int(input("Enter N: "))
i = 1
while i <= n:
  print(i)
  i += 1
```

```python
#Factorial of a number
n = int(input("Enter N: "))
fact = 1
while n > 0:
  fact *= n
  n -= 1
print("Factorial =" ,fact)
```

```python
#Fibonacci Series using while loop
n = int(input("Enter number of terms:"))
a = 0
b = 1
i = 1
while i <= n:
  print(a,end = " ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
limit = int(input("Enter limit: "))
a = 0
b = 1
while a <= limit:
    print(a, end=" ")
    a, b = b, a + b
```

```python
#Reverse of a number
n = int(input("Enter number: "))
rev = 0
while n > 0:
  digit = n % 10
  rev = rev * 10 + digit
  n = n // 10
print("Reverse number = " ,rev)
```

```python
#Check Palindrome number/string
num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num //= 10
if temp == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
#Check Palindrome number/string
s = "madam"
i , j = 0 , len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
     is_palindrome = False
     break
  i += 1
  j -= 1
if is_palindrome:
  print("Yes")
else:
  print("No")
```

```python
#check Palindrome number
st = input("Enter string: ")
rev = st[::-1]
if st == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
#Count digits in a number
num = int(input("Enter number: "))
count = 0
while num > 0:
  count += 1
  num = num//10
print("Number of digits = " ,count)

```

```python
#Search an element in a list
nums = [10, 20 ,30 , 40 , 50]
key = int(input("Enter element to search: "))

i = 0
while i < len(nums):
  if nums[i] == key:
    print("Element found at index", i)
    break
  i += 1
else:
  print("Elements not found")
```
